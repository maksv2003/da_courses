```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline


mu = 100
sigma = 10
population = np.random.normal(mu, sigma, 1000)
sample_size = 30
sample_means = []
n = 10000
for i in range (n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())
sns.displot(sample_means, kde = False)
se = population.std()/ sample_size ** 0.5
print(se)

up = population.mean() + 1.96 * se
down = population.mean() - 1.96 * se
print(up, down)


#------------------------------------------------------------------------------------#


sample_real = np.random.choice(population, sample_size, False)
se1 = sample_real.std() / sample_size ** 0.5
up1 = sample_real.mean() + 1.96 * se1
down1 = sample_real.mean() - 1.96 * se1
print('\n',se1)
print(up1, down1)
```
