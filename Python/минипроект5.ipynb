```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('/mnt/HC_Volume_18315164/home-jupyter/jupyter-m-vinokurov/transaction_data.csv')
df = df.query("transaction == 'successfull'")
df = df.groupby('name',as_index=False).agg({'transaction':'count'}).sort_values(by=['transaction'])
sns.barplot(data=df, y=df.transaction, x=df.name)
```
